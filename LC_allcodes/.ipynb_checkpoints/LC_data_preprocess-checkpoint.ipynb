## TCGA Cancer Data

```python
## TCGA_LUAD data downloaded from https://portal.gdc.cancer.gov/v1/repository?facetTab=files&filters=%7B%22op%22%3A%22and%22%2C%22content%22%3A%5B%7B%22content%22%3A%7B%22field%22%3A%22cases.case_id%22%2C%22value%22%3A%5B%22set_id%3AYKrbBd99cgJVTU8cE5YCBQ%22%5D%7D%2C%22op%22%3A%22IN%22%7D%2C%7B%22op%22%3A%22in%22%2C%22content%22%3A%7B%22field%22%3A%22files.access%22%2C%22value%22%3A%5B%22open%22%5D%7D%7D%2C%7B%22op%22%3A%22in%22%2C%22content%22%3A%7B%22field%22%3A%22files.data_category%22%2C%22value%22%3A%5B%22transcriptome%20profiling%22%5D%7D%7D%2C%7B%22op%22%3A%22in%22%2C%22content%22%3A%7B%22field%22%3A%22files.experimental_strategy%22%2C%22value%22%3A%5B%22RNA-Seq%22%5D%7D%7D%5D%7D
## Collecting lung cancer star count data with 'gene_id','gene_name','gene_type' columns from the TCGA dataset

aall = os.listdir('./LC_tcga')

for i in range(len(aall)):
    a = aall[i]
#     print(a)
    if a == 'MANIFEST.txt':
        break
        
    else:
        b = os.listdir('./LC_tcga/'+a)
        for fns in b:
            if not fns.split('.')[-1] == 'tsv':
                continue
            fname = './LC_tcga/'+a+'/'+fns

            if i == 0:
                f = pd.read_csv(fname,sep='\t',skiprows=1)
                print(f.shape)
                f = f.rename(columns={"stranded_second": i+1})
    #             print(f)
            elif i == 1:
                f1 = pd.read_csv(fname,sep='\t',skiprows=1)
                print(f1.shape)
                m_tab = pd.merge(f.loc[4:,['gene_id','gene_name','gene_type',i]], 
                                    f1.loc[4:,['gene_id','stranded_second']], left_on='gene_id', right_on='gene_id', 
                                    how='inner')
                m_tab = m_tab.rename(columns={"stranded_second": i+1})
    #             m_tableh
            else:
                f1 = pd.read_csv(fname,sep='\t',skiprows=1)
                m_tab = pd.merge(m_tab,f1.loc[4:,['gene_id','stranded_second']], left_on='gene_id', 
                                 right_on='gene_id',how='inner')
                m_tab = m_tab.rename(columns={"stranded_second": i+1})
```

```python
m_tab
```

```python
## selecting only "protein_coding" genes
can_tab = m_tab[m_tab["gene_type"]=="protein_coding"]
can_tab.shape
```

```python
## removing duplicate genes
c_unique = can_tab.drop_duplicates(subset='gene_name',keep=False)
c_unique = c_unique.reset_index()
c_unique = c_unique.drop(columns=['index'])
c_unique
```

```python
#keeping genes with <1% zero expression values 

c_shape = c_unique.shape[1]-3
c_th = int(np.floor(c_shape)*0.01)
print(c_th)

cc = (c_unique == 0).astype(int).sum(axis=1)
index_cc = np.where(cc>=c_th)
c_sel = c_unique.drop(index_cc[0])
c_sel = c_sel.reset_index()
c_sel = c_sel.drop(columns=['index'])
c_sel
```

## GTEx Healthy control data

```python
## Collecting the count data
## Healthy controls count data downloaded from https://www.gtexportal.org/home/downloads/adult-gtex/bulk_tissue_expression

fname_h = './gene_reads_lung.gct'
gtex_rc = pd.read_csv(fname_h,sep='\t',skiprows=2)
gtex_rc
```

```python
## removing duplicate genes

hc_unique = gtex_rc.drop_duplicates(subset='Description',keep=False)
hc_unique = hc_unique.rename(columns={"Description": "gene_name"})
hc_unique = hc_unique.reset_index()
hc_unique = hc_unique.drop(columns=['index'])
hc_unique
```

```python
#keeping genes with <1% zero expression values 

h_shape = hc_unique.shape[1]-3
hc_th = int(np.floor(h_shape)*0.01)
print(hc_th)

hcc = (hc_unique == 0).astype(int).sum(axis=1)
index_hcc = np.where(hcc>=hc_th)
hc_sel = hc_unique.drop(index_hcc[0])
hc_sel = hc_sel.reset_index()
hc_sel = hc_sel.drop(columns=['index'])
hc_sel
```

## Differential Expression Analysis

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
# from pydeseq2.utils import load_example_data
```

```python
dds = DeseqDataSet(
    counts=tab_altered,
    metadata=clinical_df,
    design_factors="condition",
)
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds)
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
res
```

```python
#res.to_pickle("./lc_deseq2genes.pkl")
```

```python
# Differrentially expressed genes
sigs = res[(res.pvalue < 0.01) & (abs(res.log2FoldChange>1))]
sigs
```

```python
## sorting in descending order based on log2foldchange
sind = np.flip(np.argsort(np.abs(sigs['log2FoldChange'])))
sigs = sigs.iloc[sind]
sigs
```

```python
tst = res[(res.pvalue < 0.01) & (abs(res.log2FoldChange>np.percentile(sigs.iloc[:,1],98)))]
sind1 = np.flip(np.argsort(np.abs(tst['log2FoldChange'])))
sigs1 = tst.iloc[sind1]
print(sigs1.shape)
print(sigs1)
```

```python
tab_sel = pd.merge(sigs, tab, left_on=sigs.index, right_on='gene_name', how='inner')
tab_sel
```

## Finding Gene Length 

```python
## file downloaded from Genome Reference Consortium Human Reference 38 
## (https://ftp.ncbi.nlm.nih.gov/refseq/H_sapiens/annotation/GRCh38_latest/refseq_identifiers/)

geneLenDat = []

with open('./GRCh38_latest_genomic.gff') as f:
    while True:
        lnn = f.readline().strip()
        if lnn == '':
            break
        
        if lnn[0] == '#':
            continue
        else:
#             print(lnn)
            splt = lnn.split('\t')
            snn = int(splt[3])
            enn = int(splt[4])
            IDtxt = splt[8].split(';')[0].split('=')[1]
#             print(snn,enn,IDtxt)
            if IDtxt[:4] == 'gene':
#                 print(IDtxt[5:])
                geneLenDat.append([IDtxt[5:],enn-snn+1])
```

```python
gene_len = pd.DataFrame(geneLenDat,columns=['gene_name','length'])
pd.to_pickle(gene_len,"./gene_length.pkl")
```

## exRNA data

```python
gDat_l = []
subDat_l = []

with open('../GSE106817_family.soft') as f: ## data downloaded from https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE106817
    while True:
        lnn = f.readline().strip()
        if lnn == '':
            break
        if lnn[:27] == "!Sample_title = Lung Cancer":
            fn = next(f)
            fn1 = fn.split("= ")[-1].split("\n")[0]
            a = os.listdir('../GSE106817_RAW')
            for fns in a:
                #print(fns[:9])
                if not fns.split('.')[-1] == 'txt':
                    continue
                #print(fns[:10])
                if fns[:10] == fn1:
                    fn2 = "../GSE106817_RAW/"+fns
                    fn3 = pd.read_csv(fn2,sep='\t',skiprows=7)
                    
                    gnms = list(fn3.loc[:,'G_Name'])
                    idxs = []
                    for i in range(len(gnms)):
                        s_gnms = gnms[i].split(',')
                        if len(s_gnms) == 1:
                            if s_gnms[0][0:3] == 'hsa':
                                idxs.append(i)
                    f_subDat = fn3.loc[idxs,['G_Name','G_ID','635nm','635nm.1']]
                    f_subDat['c_sub'] = f_subDat['635nm'] - f_subDat['635nm.1']
#                     f_subDat = f_subDat[f_subDat['subtract'] >= 0]
#                     f_subDat['subtract'] = np.log2(f_subDat['subtract'])
                    f_subDat = f_subDat.reset_index(drop=True)
                    gDat_l.append(f_subDat)
                    subDat_l.append(fns)
#             break
```

```python
base_gDat = gDat_l[0].loc[:,['G_Name','c_sub']]

for i in range(1,len(gDat_l)):

    add_frame = gDat_l[i].loc[:,['G_Name','c_sub']]

    base_gDat = base_gDat.merge(add_frame,on='G_Name',suffixes=(None,'_'+str(i)))
    
clst = list(base_gDat.columns)
clst[1] = 'c_sub'#+str('_1')
base_gDat.columns = clst
    
# base_gDat = base_gDat.fillna(0.1)
base_gDat
```

```python
gDat_l_nc = []
subDat_l_nc = []

with open('../GSE106817_family.soft') as f:
    while True:
        lnn = f.readline().strip()
        if lnn == '':
            break
        if lnn[:26] == "!Sample_title = non-Cancer":
            fn = next(f)
            fn1 = fn.split("= ")[-1].split("\n")[0]
            a = os.listdir('../GSE106817_RAW')
            for fns in a:
                #print(fns[:9])
                if not fns.split('.')[-1] == 'txt':
                    continue
                #print(fns[:10])
                if fns[:10] == fn1:
                    fn2 = "../GSE106817_RAW/"+fns
                    fn3 = pd.read_csv(fn2,sep='\t',skiprows=7)
                    
                    gnms = list(fn3.loc[:,'G_Name'])
                    idxs = []
                    for i in range(len(gnms)):
                        s_gnms = gnms[i].split(',')
                        if len(s_gnms) == 1:
                            if s_gnms[0][0:3] == 'hsa':
                                idxs.append(i)
                    f_subDat = fn3.loc[idxs,['G_Name','G_ID','635nm','635nm.1']]
                    f_subDat['nc_sub'] = f_subDat['635nm'] - f_subDat['635nm.1']
#                     f_subDat = f_subDat[f_subDat['subtract'] >= 0]
#                     f_subDat['subtract'] = np.log2(f_subDat['subtract'])
                    f_subDat = f_subDat.reset_index(drop=True)
                    gDat_l_nc.append(f_subDat)
                    subDat_l_nc.append(fns)
#             break
```

```python
base_gDat_nc = gDat_l_nc[0].loc[:,['G_Name','nc_sub']]

for i in range(1,len(gDat_l_nc)):

    add_frame = gDat_l_nc[i].loc[:,['G_Name','nc_sub']]
#     print(add_frame.shape)
    base_gDat_nc = base_gDat_nc.merge(add_frame,on='G_Name',suffixes=(None,'_'+str(i)))
    
# base_gDat_nc = base_gDat_nc.fillna(0.1)
base_gDat_nc
```
